```python
query = """

SELECT 
    ROUND(SUM(total_revenue),2) AS Total_Revenue
FROM customer360;

"""


pd.read_sql(
    query,
    connection
)
```

```python
query = """

SELECT
    purchase_year,
    purchase_month,
    ROUND(SUM(total_revenue),2) AS Revenue

FROM customer360

GROUP BY 
    purchase_year,
    purchase_month

ORDER BY 
    purchase_year,
    purchase_month;

"""


pd.read_sql(
    query,
    connection
)
```

```python
query = """

SELECT
    customer_unique_id,
    ROUND(SUM(total_revenue),2) AS Customer_Value

FROM customer360

GROUP BY customer_unique_id

ORDER BY Customer_Value DESC

LIMIT 10;

"""


pd.read_sql(
    query,
    connection
)
```

```python
query = """

SELECT

    s.Customer_Segment,

    COUNT(
        DISTINCT c.customer_unique_id
    ) AS Total_Customers,


    ROUND(
        SUM(c.total_revenue),
        2
    ) AS Revenue


FROM customer360 c


JOIN customer_segments s

ON c.customer_unique_id =
   s.customer_unique_id


GROUP BY
    s.Customer_Segment


ORDER BY
    Revenue DESC;


"""


pd.read_sql(
    query,
    connection
)
```

```python
query = """

SELECT

customer_unique_id,

Total_Revenue,


RANK()
OVER(
ORDER BY Total_Revenue DESC
)
AS Customer_Rank


FROM
(

SELECT

customer_unique_id,

ROUND(
SUM(total_revenue),
2
)
AS Total_Revenue


FROM customer360


GROUP BY customer_unique_id

)


LIMIT 10;


"""


pd.read_sql(
    query,
    connection
)
```

```python
query = """

WITH CustomerRevenue AS
(

SELECT

customer_unique_id,

SUM(total_revenue)
AS Revenue

FROM customer360

GROUP BY customer_unique_id

)


SELECT *

FROM CustomerRevenue


WHERE Revenue >
(
SELECT AVG(Revenue)
FROM CustomerRevenue
)


ORDER BY Revenue DESC;


"""


pd.read_sql(
    query, 
    connection
).head(10)
```
